```python
import scion
import time
import math
```

```python
def function1( x ) :
    
    return x

def function2( x ) :
    
    return ( 1 - x**2 )**10
```

```python
repeat = 1000000

scion_test = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function1, order = 10 )

print( 'scion :', scion_test.coefficients )

start = time.perf_counter()
for i in range( repeat ) : scion_test = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function1, order = 10 )
end = time.perf_counter()
time_scion = end - start

print( '-------------------------' )
print( 'scion elapsed time = {:.3} s'.format( time_scion ) )
print( '-------------------------' )
```

```python
repeat = 1000000

scion_test = scion.math.ChebyshevApproximation( lower = -2, upper = 4, function = function1, order = 10 )

print( 'scion :', scion_test.coefficients )

start = time.perf_counter()
for i in range( repeat ) : scion_test = scion.math.ChebyshevApproximation( lower = -2, upper = 4, function = function1, order = 10 )
end = time.perf_counter()
time_scion = end - start

print( '-------------------------' )
print( 'scion elapsed time = {:.3} s'.format( time_scion ) )
print( '-------------------------' )
```

```python
import matplotlib.pyplot as plot

scion_test1 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 10 )
scion_test2 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 15 )
scion_test3 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 20 )

x = [ -1. + 0.01 * i for i in range( 0, 201 ) ]
f = [ function2( value ) for value in x ]
a1 = [ scion_test1( value ) for value in x ]
a2 = [ scion_test2( value ) for value in x ]
a3 = [ scion_test3( value ) for value in x ]

# plot the data\n
plot.figure()
plot.plot( x, f, label = 'real function', color = 'red', linewidth = 1.0 )
plot.plot( x, a1, label = 'approx function order 10', color = 'green', linewidth = 1.0 )
plot.plot( x, a2, label = 'approx function order 15', color = 'blue', linewidth = 1.0 )
plot.plot( x, a3, label = 'approx function order 20', color = 'orange', linewidth = 1.0 )
plot.xlabel( 'x' )
plot.ylabel( 'y' )
plot.legend()
plot.show()
```

```python
import matplotlib.pyplot as plot

scion_test1 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 10 )
scion_test2 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 15 )
scion_test3 = scion.math.ChebyshevApproximation( lower = -1, upper = 1, function = function2, order = 20 )

x = [ -1. + 0.01 * i for i in range( 0, 201 ) ]
diff1 = [ function2( value ) - scion_test1( value ) for value in x ]
diff2 = [ function2( value ) - scion_test2( value ) for value in x ]
diff3 = [ function2( value ) - scion_test3( value ) for value in x ]

# plot the data\n
plot.figure()
plot.plot( x, diff1, label = 'approx function order 10', color = 'green', linewidth = 1.0 )
plot.xlabel( 'x' )
plot.ylabel( 'difference' )
plot.legend()
plot.show()

plot.figure()
plot.plot( x, diff2, label = 'approx function order 15', color = 'blue', linewidth = 1.0 )
plot.xlabel( 'x' )
plot.ylabel( 'difference' )
plot.legend()
plot.show()

plot.figure()
plot.plot( x, diff3, label = 'approx function order 20', color = 'orange', linewidth = 1.0 )
plot.xlabel( 'x' )
plot.ylabel( 'difference' )
plot.legend()
plot.show()
```
