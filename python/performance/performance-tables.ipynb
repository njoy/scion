```python
import scion
import time
```

```python
repeat = 10000000

x = [ i for i in range( 1, 100 ) ]
y = [ i for i in reversed( x ) ]
boundaries = [ len( x ) - 1 ]

scion_test_linlin = scion.math.LinearLinearTable( x, y )
scion_test_linlog = scion.math.LinearLogTable( x, y )
scion_test_loglin = scion.math.LogLinearTable( x, y )
scion_test_loglog = scion.math.LogLogTable( x, y )
scion_test_interpolationtable_linlin = scion.math.InterpolationTable( x, y )
scion_test_interpolationtable_linlog = scion.math.InterpolationTable( x, y, scion.interpolation.InterpolationType.LinearLog )
scion_test_interpolationtable_loglin = scion.math.InterpolationTable( x, y, scion.interpolation.InterpolationType.LogLinear )
scion_test_interpolationtable_loglog = scion.math.InterpolationTable( x, y, scion.interpolation.InterpolationType.LogLog )

print( 'scion :', scion_test_linlin( 25. ) )
print( 'scion :', scion_test_linlog( 25. ) )
print( 'scion :', scion_test_loglin( 25. ) )
print( 'scion :', scion_test_loglog( 25. ) )
print( 'scion :', scion_test_interpolationtable_linlin( 25. ) )
print( 'scion :', scion_test_interpolationtable_loglin( 25. ) )
print( 'scion :', scion_test_interpolationtable_linlog( 25. ) )
print( 'scion :', scion_test_interpolationtable_loglog( 25. ) )

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_linlin( 25. )
end = time.perf_counter()
time_scion_linlin = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_linlog( 25. )
end = time.perf_counter()
time_scion_linlog = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_loglin( 25. )
end = time.perf_counter()
time_scion_loglin = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_loglog( 25. )
end = time.perf_counter()
time_scion_loglog = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_interpolationtable_linlin( 25. )
end = time.perf_counter()
time_scion_interpolationtable_linlin = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_interpolationtable_linlog( 25. )
end = time.perf_counter()
time_scion_interpolationtable_linlog = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_interpolationtable_loglin( 25. )
end = time.perf_counter()
time_scion_interpolationtable_loglin = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test_interpolationtable_loglog( 25. )
end = time.perf_counter()
time_scion_interpolationtable_loglog = end - start

print( '-------------------------' )
print( 'scion lin-lin elapsed time = {:.3} s'.format( time_scion_linlin ) )
print( 'scion lin-log elapsed time = {:.3} s'.format( time_scion_linlog ) )
print( 'scion log-lin elapsed time = {:.3} s'.format( time_scion_loglin ) )
print( 'scion log-log elapsed time = {:.3} s'.format( time_scion_loglog ) )
print( '-------------------------' )
print( 'scion interpolation table single lin-lin elapsed time = {:.3} s'.format( time_scion_interpolationtable_linlin ) )
print( 'scion interpolation table single lin-log elapsed time = {:.3} s'.format( time_scion_interpolationtable_linlog ) )
print( 'scion interpolation table single log-lin elapsed time = {:.3} s'.format( time_scion_interpolationtable_loglin ) )
print( 'scion interpolation table single log-log elapsed time = {:.3} s'.format( time_scion_interpolationtable_loglog ) )
print( '-------------------------' )
```

```python
repeat = 10000000

x = [ i for i in range( 1, 100 ) ]
y = [ i for i in reversed( x ) ]
boundaries = [ 20, 25, 75, len( x ) - 1 ]
interpolants = [ scion.interpolation.InterpolationType.LinearLinear, 
                 scion.interpolation.InterpolationType.LinearLog, 
                 scion.interpolation.InterpolationType.LogLinear, 
                 scion.interpolation.InterpolationType.LogLog ]

scion_test = scion.math.InterpolationTable( x, y, boundaries, interpolants )

print( 'scion :', scion_test( 14.5 ) )
print( 'scion :', scion_test( 22.1 ) )
print( 'scion :', scion_test( 74.5 ) )
print( 'scion :', scion_test( 75.3 ) )

start = time.perf_counter()
for i in range( repeat ) : result = scion_test( 14.5 )
end = time.perf_counter()
time_scion_1 = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test( 22.1 )
end = time.perf_counter()
time_scion_2 = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test( 74.5 )
end = time.perf_counter()
time_scion_3 = end - start

start = time.perf_counter()
for i in range( repeat ) : result = scion_test( 75.3 )
end = time.perf_counter()
time_scion_4 = end - start

print( '-------------------------' )
print( 'scion evaluate 1 elapsed time = {:.3} s'.format( time_scion_1 ) )
print( 'scion evaluate 2 elapsed time = {:.3} s'.format( time_scion_2 ) )
print( 'scion evaluate 3 elapsed time = {:.3} s'.format( time_scion_3 ) )
print( 'scion evaluate 4 elapsed time = {:.3} s'.format( time_scion_4 ) )
print( '-------------------------' )
```
